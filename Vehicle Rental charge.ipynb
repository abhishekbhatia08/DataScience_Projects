```python
class vechile:
    def __init__(self,vechile=None):
        self.count=0
        
        if vechile=='car':
            self.vechile=vechile
                        

            
        elif vechile=='van':
            self.vechile=vechile
                        

                        

            
        elif vechile=='moving_truck':
            self.vechile=vechile
        
            
            
    def reserve(self):
        res = input('Would you like me to reserve the vehicle for you(y/n): ')
        if res == 'y':
            print(self.vechile,'Reserved!')
            self.count=self.count+1
        else:
            print('Ok then!')        
    def rent_charge(self):
        time = int(input('Enter the number of hours you want to rent for: '))
        if(self.vechile=='car'):
            price=500
        elif(self.vechile=='van'):
            price=700
        elif(self.vechile=='moving_truck'):
            price=800
        rental=price*time
        print("rental charge is",rental)
       
    def check(self):
        if(self.count<5):
            print("available",(5-self.count),self.vechile)
            
        else:
            print("not available")
    def cancel(self):
        res = input('Are you sure you want to cancel(y/n): ')
        if res =='y':
            print('Cancelled!')
            self.count=self.count-1
            
        else:
            print('Ok then!')         
            
book=input("would you like to book y/n")     
while(book=='y'):
    v=input("enter the vechile name")
    v1=vechile(v)  
    v1.reserve()

    v1.rent_charge()
    v1.check()
    book=input("would you like to book y/n") 
    



            
```
